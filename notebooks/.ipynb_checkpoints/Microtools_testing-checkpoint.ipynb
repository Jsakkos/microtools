```python
import sys
sys.path.append('/content/drive/MyDrive/microtools/Scripts')
import Scripts.imgutils as imgutils
import matplotlib.pyplot as plt
# import the libraries mentioned above
import czifile as zis
import xmltodict
import os
import numpy as np
#import xml.etree.ElementTree as ET
from IPython.display import display, HTML
```

```python
file = './Images/Control-02.czi'
image = zis.imread(file)
img = np.squeeze(image)[1,:,:]
plt.figure(figsize=(4,4))
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
md = imgutils.get_metadata_czi(file, dim2none=False)

def find_chan(metadata,channel):
    for i, dic in enumerate(metadata['Channels']):
        #print(i, dic)
        if dic == channel:
            return i
    return -1
DsRed = imgutils.find_chan(md,'DsRed')
img = np.squeeze(image)
```

```python
from cellpose import models, utils,plot,io

use_GPU = models.use_gpu()
print('>>> GPU activated? %d'%use_GPU)
```

```python
model = models.Cellpose(gpu=True, model_type='cyto')
#enter  channel order/position
dsred_channel = imgutils.find_chan(md,'DsRed')
mng_channel = 0

folder = '/content/drive/MyDrive/microtools/Images/'
for f in Path(folder).rglob('*.czi'):
    if os.path.exists(str(f.stem) + '_seg.npy'):
      continue
    else:
      print(f)
      image = zis.imread(f)
      img = np.squeeze(image)
      dsred = img[dsred_channel,:,:]
      masks, flows, styles, diams = model.eval(dsred, diameter=20)
      # save results
      io.masks_flows_to_seg(dsred, masks, flows, diams, f)
      fig = plt.figure(figsize=(12,5))
      plot.show_segmentation(fig, dsred, masks, flows[0])
      plt.tight_layout()
      plt.show()
```
